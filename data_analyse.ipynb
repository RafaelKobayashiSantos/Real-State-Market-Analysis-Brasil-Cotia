```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dados_imoveis.csv")

df["preco_m2"] = df["preco"] / df["area"].replace(".", ",").astype(float)

df["bairro"] = (
    df["descricao"]
    .str.split(" em ")
    .str[-1]
    .str.split(",")
    .str[0]
)

df["cidade"] = (
    df["descricao"]
    .str.split(",")
    .str[-1]
    .str.strip()
)

df["vagas"] = df["vagas"].fillna(0).astype(int)

df.to_excel( "dados_imoveis_pby.xlsx", index=False )
```
